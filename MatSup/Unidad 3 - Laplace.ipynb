## Recordando Fourier

Vamos a recordar un poco la transformada de Fourier:

$F({\omega}) = {\displaystyle \int_{-\infty}^{\infty} f(t){\bullet} {e}^{-j{ω}t}dt} $

Recordar: $F({\omega}): \mathbb{R} {\to} \mathbb{C}$ (función compleja de variable real)

## Introducción

Lo que va a proponer Laplace es una nueva transformada que permita calcularse para aquellas funciones que no sean de **modulo integrable** bajo ciertas condiciones: agregar un termino a la transformada propuesta por Fourier para que exista la integral en todo el dominio

$F({\omega}, a) = {\displaystyle \int_{-\infty}^{\infty} f(t){\bullet} {e}^{-j{ω}t}{\bullet}\color{red}{{e}^{-at}}dt}$

Reacomodando los términos se llega a:

$F(s) = {\displaystyle \int_{-\infty}^{\infty} f(t){\bullet} {e}^{-st}}dt$

Nosotros vamos a estar viendo la transformada de Laplace **unilateral derecha**

$F(s) = {\displaystyle \int_0^{\infty} f(t){\bullet} {e}^{-st}}dt$ donde $Re(s) = a{\geq}0$ para que sea exponencial decreciente

$F(s): \mathbb{C} {\to} \mathbb{C}$ (función compleja de variable compleja)

## Teorema de Existencia

Sea $f(t)$ una función **seccionalmente continua** en todo intervalo [0;N] y $f(t)$ de **orden exponencial** γ para t > N ${\to}$ ${∃}$ $\mathcal{L}[f(t)]$


### Qué es una función de orden exponencial?

La función $f(t)$ es de orden exponencial γ $\leftrightarrow$ ∃ M,γ ∈ $\mathbb{R}$ $/$ $∀$ $t>N: |e^{-γt}f(t)|$< M si t ${\to}$ $∞$

Observación: la condición  $|e^{-γt}f(t)|< M$ $≡$ $|f(t)|< Me^{γt}$ (funcion f(t) en módulo no puede crecer más que $Me^{-γt}$)

## Propiedades de la Transformada de Laplace



*   Linealidad

*   Primera propiedad de traslación

*   Segunda propiedad de traslación

*   Cambio de escala

*   Transformada de las derivadas

*   Transformada de las integrales

*   Multiplicacion por $t^n$

*   División por $t$

*   Teorema del valor inicial

*   Teorema del valor final

### Linealidad

Si $F(s)$ = $\mathcal{L}[f(t)]$ ∧ $\mathcal{L}[g(t)]$ ∧ $k_1$,$k_2$ ∈ $\mathbb{R}$ ${\to}$ $\mathcal{L}[k_1f(t) \pm k_2g(t)] = k_1F(s)\pm k_2G(s)$

### Primera propiedad de Traslación

Si $F(s)$ = $\mathcal{L}[f(t)]$ ${\to}$ $\mathcal{L}[e^{at}f(t)] = F(s-a)$

Esto quiere decir que multiplicar por $e^{at}$ en el tiempo equivale a efectuar una traslación de "a" unidades en el dominio transformado de Laplace

### Segunda propiedad de Traslación

Si $F(s)$ = $\mathcal{L}[f(t)]$ ∧
$g(t) =
  \begin{cases}
    f(t-a)       & \quad \text{si } t \text{ > a}\\
    0  & \quad \text{si } t \text{ < a}
  \end{cases}
$
${\to}$ $\mathcal{L}[g(t)] = e^{-as}F(s)$

Esto significa que una traslación de "a" unidades hacia la derecha en el tiempo equivale a multiplicar por $e^{-as}$ en el dominio transformado de Laplace

### Cambio de Escala

Si $F(s)$ = $\mathcal{L}[f(t)]$ ${\to}$ $\mathcal{L}[f(at)] = \frac{1}{a}F(\frac {s}{a})$

### Transformada de Laplace de las Derivadas

Si $F(s)$ = $\mathcal{L}[f(t)]$ ${\to}$ $\mathcal{L}[f'(t)] = sF(s) - f(0)$

Generalización para derivadas de orden superior:

$\mathcal{L}[f''(t)] = s^2F(s) - sf(0) - f'(0)$

$\mathcal{L}[f'''(t)] = s^3F(s) - s^2f(0) - sf'(0) - f''(0)$

### Transformada de Laplace de las Integrales

Si $F(s)$ = $\mathcal{L}[f(t)]$ ${\to}$ $\mathcal{L}[{\displaystyle \int_0^{t} f(u)du]} = \frac{F(s)}{s}$

### Multiplicación por $t^n$

Si $F(s)$ = $\mathcal{L}[f(t)]$ ${\to}$ $\mathcal{L}[t^nf(t)] = (-1)^nF^{(n)}(s)$

### División por $t$

Si $F(s)$ = $\mathcal{L}[f(t)]$ ${\to}$ $\mathcal{L}[\frac{f(t)}{t}] = {\displaystyle \int_s^{∞} F(u)du}$ **SIEMPRE Y CUANDO** $∃$ $\lim\limits_{t \to 0} \frac{f(t)}{t}$

### Teorema del Valor Incial

Si existen los límites indicados, se cumple: $\lim\limits_{t \to 0} f(t)$ = $\lim\limits_{s \to ∞} sF(s)$

### Teorema del Valor Final

Si existen los límites indicados, se cumple: $\lim\limits_{t \to ∞} f(t)$ = $\lim\limits_{s \to 0} sF(s)$

## Evaluación de Integrales por Transformada de Laplace

La Transformada de Laplace nos va a permitir calcular en forma rápida resultados numéricos de integrales impropias, cuya resolución analítica sería extensa, tediosa, e incluso a veces imposible (por no existir primitiva)

Si $F(s)$ = $\mathcal{L}[f(t)]$ entonces por definición se cumple que ${\displaystyle \int_0^{\infty} f(t){\bullet} {e}^{-st}}dt = F(s)$

Tomando limite para s ${\to}$0, es: ${\displaystyle \int_0^{\infty} f(t)dt} = F(0)$ siempre que la integral converja

Generalizando:

Tomando limite para s ${\to}$a, es: ${\displaystyle \int_a^{\infty} f(t)dt} = F(a)$ siempre que la integral converja

## Funciones Especiales

### Función Impulso Unitario / Delta de Dirac

Muchas veces, nos vemos en la necesidad de representar una situación en la cual hay un valor de
algo en un lapso de tiempo muy pequeño, por ejemplo si aplicamos en un instante dado una fuerza
a un bloque de masa, y enseguida lo soltamos para ver como responde.
¿Es eso una función? Solamente tendría un valor en t=0 y en el resto sería nula.

Consideremos el siguiente rectángulo de área unitaria:

Impulso Unitario

${\displaystyle \int_0^{τ} \frac{1}{τ} dτ} = 1$ ${\to}$ Por eso es unitario

A menor τ, más chica es la base y más grande la altura. Si hacemos tender τ ${\to}0$, la altura tiende a $∞$

Impulso

Definimos la función Delta o impulso unitario $δ(t) = \lim\limits_{t \to 0} F_τ(t)$

siendo $F_τ(t) =
  \begin{cases}
    \frac{1}{τ}       & \quad \text{si } t \text{ 0 $\leq$ t < τ }\\
    0  & \quad \text{si } t \text{ > 0}
  \end{cases}
$

La transformada de Laplace del impulso unitario es:

$\mathcal{L}[δ(t)] = 1$

### Función Escalón Unitario / Heaviside

Esta función es muy importante, ya que multiplicada por alguna constante, puede representar
funciones constantes a partir de un instante dado, como por ejemplo, una señal de tensión
continua en un circuito eléctrico, una fuerza constante aplicada sobre algún sistema mecánico, etc.

$E(t) =
  \begin{cases}
    1       & \quad \text{si } t \text{ > 0}\\
    0  & \quad \text{si } t \text{ < 0}
  \end{cases}
$

$E(t-a) =
  \begin{cases}
    1       & \quad \text{si } t \text{ > a}\\
    0  & \quad \text{si } t \text{ < a}
  \end{cases}
$

Escalon

La transformada de Laplace del Escalón Unitario es:

$\mathcal{L}[E(t-a)] = \frac{e^{-sa}}{s}$

Observación: también podría haberse calculado aplicando la segunda propiedad de la traslación.

Utilizando funciones escalones desplazados podemos dar funciones que tienen
diferentes expresiones en distintos intervalos, pero escribiendo en un solo renglón, y ello facilita el
cálculo de la Transformada de Laplace.

## Teorema de Convolución o Teorema de Borel

Si $f(t)= \mathcal{L}^{-1}[F(s)]$ ∧ $g(t)= \mathcal{L}^{-1}[G(s)]$ ${\to}$ $\mathcal{L}^{-1}[F(s){\bullet}G(s)] = {\displaystyle \int_0^t f(u)g(t-u)}du$

Es una propiedad que nos ofrece una forma de antitransformar el producto de dos funciones en el dominio "s".
El resultado de la integral nos queda en función de "t" al aplicar Barrow.

<h2>Funciones Principales</h2>

```python
import sympy as smp
from sympy.abc import s,t,x,y,z, a, u, U, k, X, Y, Z
from sympy.integrals import laplace_transform
from sympy.integrals import inverse_laplace_transform
from sympy import Heaviside
from sympy import Eq, sympify, limit
from sympy import symbols, Function, solve, integrate, Integral, Matrix, solve_linear_system


def L(f):
    return laplace_transform(f, t, s, noconds=True)


def inv_L(F, variable_retorno):
    return inverse_laplace_transform(F, s, variable_retorno, noconds=True)


def L_diff(f, f0, fp0, fpp0=0, orden_derivada=1):  # esta función es para calcular transformadas de derivadas
    """
    input:
    f es una variable simbólica de sympy que representa a la función f(t) a la que se quiere calcular
      la transformada de alguna de sus derivadas.

    n puede asumir los valores de 0,1,2,3 y representa el orden de la derivada de la función f(t)
    """
    F = str(f).upper().split('(')[0]
    F = symbols(F)
    F = Function(F)(s)
    if orden_derivada == 0:
        L = F
    elif orden_derivada == 1:
        L = s * F - f0
    elif orden_derivada == 2:
        L = s ** 2 * F - s * f0 - fp0
    elif orden_derivada == 3:
        L = s ** 3 * F - s ** 2 * f0, -s * fp0 - fpp0
    return L

def armar_ecuacion(primer_miembro, segundo_miembro):
    return Eq(primer_miembro, segundo_miembro)

def resolver_ecuacion(ecuacion, incognita):
    return solve([ecuacion], (incognita))

def resolver_sistema_ecuaciones(ecuaciones, incognita):
    return solve(ecuaciones, (incognita))

def transformada_division_por_t(funcion):
    return smp.integrate(funcion, (U,s,limit(funcion, s, smp.oo)))

def mostrar_resultados(resultados):
    for i in range(0, len(resultados)):
      display(resultados[i])
      print("\n")


x = Function('x')(t) # Definimos a x, y, z como funciones del tiempo
y = Function('y')(t)
z = Function('z')(t)
X = Function('X')(s) # Defnimos a X, Y, Z como funciones de s
Y = Function('Y')(s)
Z = Function('Z')(s)

t = smp.Symbol('t', positive=True)
u = smp.Symbol('u', positive=True)
```
